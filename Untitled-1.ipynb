### Comparison Of The Top 3 Streaming Platform


```python
## Group the data by streaming platform and aggregate the metrics
platform_comparison = df.agg({
    'in_spotify_playlists': 'sum',
    'in_spotify_charts': 'sum',
    'in_apple_playlists': 'sum',
    'in_apple_charts': 'sum',
    'in_deezer_playlists': 'sum',
    'in_deezer_charts': 'sum',
})

# Print the aggregated data
print(platform_comparison)
```

```python
# Convert the data to a DataFrame
platform_comparison_df = platform_comparison.reset_index()
platform_comparison_df.columns = ['metric', 'count']

# Find the platform with the maximum count for each metric
top_platforms = platform_comparison_df.groupby('metric')['count'].idxmax()
top_platforms_info = platform_comparison_df.loc[top_platforms]

# Print the top-performing platforms for each metric
print("Top-performing platforms for each metric:")
print(top_platforms_info)
```

```python
# Convert 'count' column to numeric
top_platforms_info['count'] = pd.to_numeric(top_platforms_info['count'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_platforms_info['metric'], top_platforms_info['count'], color='skyblue')
plt.xlabel('Streaming Platform')
plt.ylabel('Count')
plt.title('Top-performing platforms for each metric')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Summing the counts across all platforms for each metric
total_spotify_playlists = df['in_spotify_playlists'].sum()
total_spotify_charts = df['in_spotify_charts'].sum()
total_apple_playlists = df['in_apple_playlists'].sum()
total_apple_charts = df['in_apple_charts'].sum()
total_deezer_playlists = sum(int(value.replace(',', '')) for value in df['in_deezer_playlists'])
total_deezer_charts = df['in_deezer_charts'].sum()


totals_df = pd.DataFrame({
    'metric': ['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 
               'in_apple_charts', 'in_deezer_playlists', 'in_deezer_charts'],
    'count': [total_spotify_playlists, total_spotify_charts, total_apple_playlists, 
              total_apple_charts, total_deezer_playlists, total_deezer_charts]
})


# Plotting
plt.figure(figsize=(10, 6))
plt.bar(totals_df['metric'], totals_df['count'], color=['green', 'blue', 'orange', 'red', 'purple', 'cyan'])
plt.xlabel('Metric')
plt.ylabel('Count')
plt.title('Total counts for each streaming platform')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
